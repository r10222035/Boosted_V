```python
def read_sample_size(df):
    col = ['kappa', 'Sample', 'Train W+', 'Train W-', 'Train Z', 'Test W+',	'Test W-', 'Test Z', 'time']
    IPython.display.display(df[col])
    
    cols = ['Train W+', 'Train W-', 'Train Z', 'Test W+', 'Test W-', 'Test Z']
    for i, row in df[cols].iterrows():
        training_set = f"${row['Train W+']//1000}k+{row['Train W-']//1000}k+{row['Train Z']//1000}k$"
        testing_set = f"${row['Test W+']//1000}k+{row['Test W-']//1000}k+{row['Test Z']//1000}k$"
    
        print(f'{training_set:20} {testing_set:20}')
    

def read_training_result(df):
    pd.set_option('display.precision', 3)
    col = ['kappa', 'Sample', 'Overall ACC', 'W+ AUC', 'W+ ACC', 'W- AUC', 'W- ACC', 'Z AUC', 'Z ACC', 'time']
    IPython.display.display(df[col])
```

```python
read_sample_size(training_results_df)
```

```python
read_training_result(training_results_df)
```
