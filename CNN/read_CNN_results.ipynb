```python
def read_sample_size(df):
    col = ['kappa', 'Sample', 'Train W+', 'Train W-', 'Train Z', 'Validation W+', 'Validation W-', 'Validation Z', 'Test W+',	'Test W-', 'Test Z', 'time']
    IPython.display.display(df[col])
    
    cols = ['Train W+', 'Train W-', 'Train Z', 'Test W+', 'Validation W+', 'Validation W-', 'Validation Z', 'Test W-', 'Test Z']
    for i, row in df[cols].iterrows():
        training_set = f"${row['Train W+']//1000}k+{row['Train W-']//1000}k+{row['Train Z']//1000}k$"
        validation_set = f"${row['Validation W+']//1000}k+{row['Validation W-']//1000}k+{row['Validation Z']//1000}k$"
        testing_set = f"${row['Test W+']//1000}k+{row['Test W-']//1000}k+{row['Test Z']//1000}k$"
    
        print(f'{training_set:20} {validation_set:20} {testing_set:20}')
        print(f'{i} {training_set} & {validation_set} & {testing_set}')

def read_training_result(df):
    pd.set_option('display.precision', 3)
    col = ['kappa', 'Sample', 'Overall ACC', 'W+ AUC', 'W+ ACC', 'W- AUC', 'W- ACC', 'Z AUC', 'Z ACC', 'time']
    IPython.display.display(df[col])
```

```python
read_sample_size(training_results_df)
```

```python
read_training_result(training_results_df)
```

# Event

```python
file_name = 'event_training_results.csv'
training_results_df = pd.read_csv(file_name)
```

```python
training_results_df
```

```python
event_type = ['W+W+', 'W-W-', 'ZZ', 'W+Z', 'W-Z', 'W+W-']
    
def read_event_sample_size(df):
    col = ['kappa', 'Sample']
    col.extend([f'Train {_type}' for _type in event_type])
    col.extend([f'Validation {_type}' for _type in event_type])
    col.extend([f'Test {_type}' for _type in event_type])
    col.append('time')
    IPython.display.display(df[col])
    
    for i, row in df.iterrows():
        text = ''
        for _type in event_type:
            text += str((row[f'Train {_type}'] + row[f'Validation {_type}'] + row[f'Test {_type}']) // 1000) + 'k\t'
        print(i,text)

def read_event_training_result(df):
    pd.set_option('display.precision', 3)
    col = ['kappa', 'Sample', 'Model', 'Overall ACC',]
    for _type in event_type:
        col.append(f'{_type} AUC')
        col.append(f'{_type} ACC')
        
    col.append('time')    
    IPython.display.display(df[col])
```

```python
read_event_sample_size(training_results_df)
```

```python
read_event_training_result(training_results_df)
```
